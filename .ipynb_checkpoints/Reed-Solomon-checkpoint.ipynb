$$\text{Пример кода Рида-Соломона}$$
$$ \text{Код Рида-Соломона является кодом с максимально достижимым кодовым расстоянием (МДР), т.е. } \boxed{d = n - k + 1} $$
$$ \text{Также кодом Рида-Соломона является код БЧХ, если корни его } $$
$$ \text{порождающего многочлена принадлежат тому же полю, что и коэффициенты} \implies \boxed{n = q - 1} $$
$$ \text{Несистематическое кодирование: } V = (a(1), a(\alpha), ..., a(\alpha ^ {q - 2})) $$
$$\text{Корни порождающего многочлена } g(x) \text{ определены как } \alpha ^ {b}, \alpha ^ {b + 1}, ..., \alpha ^ {b + d - 2}$$
$$\text{Реализуем код, который исправляет 2 ошибки }(t=2) \text{, в конечном поле }GF(2 ^ 6) \text{. При этом } b = 1 $$
$$\text{Для поля } GF(2 ^ 6) \text{ имеем } p = 2, m = 6 $$
$$\text{Кодовое расстояние задается как } d = 2t + 1 = 2 \cdot 2 + 1 = 5$$
$$\text{Т.к. } b + d - 2 = 1 + 5 - 2 = 4 \text{, то корни порождающего многочлена } \alpha, \alpha ^ 2, \alpha ^ 3, \alpha ^ 4 $$
$$\text{Найдем порождающий многочлен:} $$
$$ g(x) = (x - \alpha)(x - \alpha ^ 2)(x - \alpha ^ 3)(x - \alpha ^ 4) = (x ^ 2 + (\alpha + \alpha ^ 2)x + \alpha ^ 3)(x ^ 2 + (\alpha ^ 3 + \alpha ^ 4)x + \alpha ^ 7) = $$
$$ = (x ^ 2 + \alpha ^ {57}x + \alpha ^ 3)(x ^ 2 + \alpha ^ {59}x + \alpha ^ 7) = x ^ 4 + \alpha ^ {43}x ^ 3 + \alpha ^ {9}x ^ 2 + \alpha ^ {48}x + \alpha ^ {10} $$
$$\text{Т.к. код Рида-Соломона, то } n = q - 1 = p ^ m - 1 = 63, r = \deg (g(x)) = n - k = 4 \implies k = 63 - 4 = 59 $$
$$\implies (63,59) \text{ - линейный код} $$
$$ \text{Проверим МДР: } d = n - k + 1 = 63 - 59 + 1 =5 $$
$$\implies \deg(a(x)) < 59\text{, где } a(x)\text{ - информационный полином и } $$
$$ deg(V(x)) < 63 \text{, где } V(x)\text{ - закодированное для передачи сообщение(полином)} $$
$$ \text{Причем коэффициенты информационного и закодированного сообщения } \in \mathbb{F}_{2 ^ 6} $$
$$ \text{Пусть необходимо передать следующую информацию: } $$
$$
\begin{array}{c:c:c:c:c:c:c:c:c:c}
   00 \dots 0 & 010111 & 000000 & 000000 & 101010 & 000000 & 010110 & 011011 & 000000 & 101011 \\
   \dots & \alpha ^ {18} & 0 & 0 & \alpha ^ {41} & 0 & \alpha ^ {54} & \alpha ^ {6} & 0 & \alpha ^ {24}
\end{array}
$$
$$ \implies a(x) = \alpha ^ {18} x ^ {8} + \alpha ^ {41} x ^ {5} + \alpha ^ {54} x ^ {3} + \alpha ^ {6} x ^ {2} + \alpha ^ {24}; $$
$$ \text{Воспользуемся систематическим кодированием для закодирования информационного сообщения: } $$
$$ V(x) = a(x) \cdot x ^ r + (a(x) \cdot x ^ r) \mod (g(x)) $$
$$ \text{Имеем закодированное сообщение: } V(x) = \alpha ^ {18} x ^ {12} + \alpha ^ {41} x ^ {9} + \alpha ^ {54} x ^ {7} + $$
$$ + \alpha ^ {6} x ^ {6} + \alpha ^ {24} x ^ {4} + \alpha ^ {47} x ^ {3} + \alpha ^ {56} x ^ {2} + \alpha ^ {40} x + \alpha ^ {17} $$
$$\text{Переданное по каналу связи закодированное сообщение имеет вид: } U(x) = V(x) + e(x)\text{, где } e(x) \text{ - вектор ошибки;} $$
$$\text{Проверочная матрица: } H = \begin{bmatrix}
   1 & \alpha & \alpha ^ {2} & \dots & \alpha ^ {50} \\
   1 & \alpha ^ {2} & \alpha ^ {4} & \vdots & \alpha ^ {100} \\
   1 & \alpha ^ {3} & \alpha ^ {9} & \vdots & \alpha ^ {150} \\
   1 & \alpha ^ {4} & \alpha ^ {8} & \dots & \alpha ^ {200}
\end{bmatrix} \implies 
\text{Синдром ошибки: }
S(U(x))=\begin{pmatrix}
   S_1 \\
   S_2 \\
   S_3 \\
   S_4 
\end{pmatrix}=\begin{pmatrix}
   U(\alpha) = e(\alpha) \\
   U(\alpha ^ 2) = e(\alpha ^ 2) \\
   U(\alpha ^ 3) = e(\alpha ^ 3) \\
   U(\alpha ^ 4) = e(\alpha ^ 4) 
\end{pmatrix} $$
$$\text{Допустим принятое сообщение не имеет ошибок }(U(x) = V(x) + 0 = V(x)) \implies \text{ синдром ошибки: } $$
$$ S_1 = \alpha ^ {18 + 12} + \alpha ^ {41 + 9} + \alpha ^ {54 + 7} + \alpha ^ {6 + 6} + \alpha ^ {24 + 4} + \alpha ^ {47 + 3} + \alpha ^ {56 + 2} + \alpha ^ {41} + \alpha ^ {17} = 0 $$
$$ S_2 = 0 $$
$$ S_3 = \alpha ^ {18} \cdot (\alpha ^ {3}) ^ {12} + \alpha ^ {41} \cdot (\alpha ^ {3}) ^ {9} + \alpha ^ {54} \cdot (\alpha ^ {3}) ^ {7} + \alpha ^ {6} \cdot (\alpha ^ {3}) ^ {6} + \alpha ^ {24} \cdot (\alpha ^ {3}) ^ {4} + \alpha ^ {47} \cdot (\alpha ^ {3}) ^ {3} + \alpha ^ {56} \cdot (\alpha ^ {3}) ^ {2} + \alpha ^ {40} \cdot (\alpha ^ {3}) + \alpha ^ {17} = 0  $$
$$ S_4 = 0 \implies \text{ ошибок нет } $$
$$ \large{\text{Допустим одну ошибку: }} $$
$$ U(x) = \alpha ^ {18} x ^ {12} + \underline{\alpha ^ {43} x ^ {10}} + \alpha ^ {41} x ^ {9} + \alpha ^ {54} x ^ {7} + \alpha ^ {6} x ^ {6} + \alpha ^ {24} x ^ {4} + \alpha ^ {47} x ^ {3} + \alpha ^ {56} x ^ {2} + \alpha ^ {40} x + \alpha ^ {17}; $$
$$ S_1 = \alpha ^ {43} \cdot \alpha ^ {10} = \alpha ^ {53}; $$
$$ S_2 = \alpha ^ {43} \cdot (\alpha ^ {2}) ^ {10} = 1 $$
$$ S_3 = \alpha ^ {43} \cdot (\alpha ^ {3}) ^ {10} = \alpha ^ {43} \cdot \alpha ^ {30} = \alpha ^ {10}; $$
$$ S_4 = \alpha ^ {43} \cdot (\alpha ^ {4}) ^ {10} = \alpha ^ {43} \cdot \alpha ^ {40} = \alpha ^ {20}; $$
$$ \underline{\text{Алгоритм Евклида: }} $$
$$ \text{Начальные значения: }a_0 = x ^ {2t} = x ^ 4, a_1 = b = S(x) = \displaystyle\sum_{j = 1} ^ {2t} S _ j x ^ {j - 1} = \alpha ^ {20} x ^ {3} + \alpha ^ {10} x ^ {2} + x + \alpha ^ {53}, y_0 = 0, y_1 = 1 $$
$$ \text{Формулы: } a_i = a_{i+1}q_{i+1} + a_{i+2}, y_{i+2} = y_i - y_{i+1} \cdot q_{i+1} $$
$$ \text{Найдем }y_i\text{ , где }\deg(a_i)\leqslant t-1 = 1 $$
$$ \text{При }i = 0\text{ : } $$
$$ a_0 = a_1 \cdot q_1 + a_2 \implies q_1 = (\alpha ^ {4} + \alpha ^ {3} + \alpha ^ {2} + \alpha)x + (\alpha ^ {5} + \alpha ^ {4} + \alpha ^ {3} + \alpha ^ {2} + \alpha + 1);$$
$$ a_2 = \alpha ^ {5} + \alpha ^ {4} + \alpha ^ {3} = \alpha ^ {23}; $$
$$ \deg(a_2) \leqslant t - 1 $$
$$ 0 \leqslant 1 \text{ - да } \implies i = 2 $$
$$ y_2 = y_0 - y_1 \cdot q_1 = 0 - 1 (\alpha ^ {43}x + \alpha ^ {33}) = \alpha ^ {43}x + \alpha ^ {33}; $$
$$ \text{Найдем многочлен локаторов ошибок: } \delta(x) = \chi y_{i} = \chi y_{2} \text{ при условии, что } \delta(0) = 1\text { т.е подберем } \chi $$
$$ \chi = \alpha ^ {30} \text{ т.к. } \alpha ^ {30} (\alpha ^ {43} \cdot 0 + \alpha ^ {33}) = 1  $$
$$ \implies \delta(x) = \alpha ^ {30} (\alpha ^ {43}x + \alpha ^ {33}) = \alpha ^ {10}x + 1  $$
$$ \text{Корни многочлена локаторов ошибок равны обратным величинам локаторов ошибок} $$
$$ X_i = (\alpha ^ {j_1}) ^ {-1} \text{ - корень многочлена локаторов ошибок, а } j_i \text{ - место ошибки}$$
$$ \text{Корень } X_1 = \alpha ^ {53} \implies \text{ локатор ошибок равен } (\alpha ^ {53}) ^ {-1} = \alpha ^ {10} \implies j_1 = 10 \implies \text{ошибка в 10м месте} $$
$$ \text{Многочлен значений значений ошибок имеет вид:} $$
$$ \Omega(x) = (-1) ^ {i} \chi a_i = (-1) ^ {i} \chi a_2 = $$
$$ = (-1) ^ 2 \cdot \alpha ^ {30} \cdot \alpha ^ {23} = \alpha ^ {53}; $$
$$ \text{Производная многочлена локаторов ошибок } \delta'(x) = \alpha ^ {10} $$
$$ \text{Теперь по алгоритму Форни получаем значения для вектора ошибок:} $$
$$ Y_i = - \frac{\Omega(X_i)}{\delta'(X_i)} $$
$$ Y_1 = - \frac{\Omega(X_1)}{\delta'(X_1)} = - \frac{\alpha ^ {53}}{\alpha ^ {10}} = \alpha ^ {53} \cdot (\alpha ^ {10}) ^ {-1} = $$
$$ = \alpha ^ {53} \cdot \alpha ^ {53} = \alpha ^ {43}  $$
$$ \implies \text{ вектор ошибок имеет вид } e(x) = \alpha ^ {43} x ^ {10}; $$
$$ \implies \text{ правильная комбинация определяется как } V(x) = U(x) + e(x); $$
$$ \underline{\text{Алгоритм Берлекэмпа-Месси: }} $$
$$ \text{Начальные условия: } r = 0, L = 0, \delta(x) = 1, B(x) = 1; $$
$$ \text{Итерация } r = r + 1 = 0 + 1 = 1 $$
$$ \Delta_1 = \displaystyle\sum_{j = 0} ^ {L = 0} \delta_j S_{r - j} = \delta_0 S_{1}  = \alpha ^ {53}; $$
$$ \Delta_1 = \alpha ^ {53} \neq 0 \implies T(x) = \delta(x) + \Delta_1 x B(x) = 1 + \alpha ^ {53} \cdot x = \alpha ^ {53} x + 1; $$
$$ 2 \cdot 0 \leqslant 1 - 1 \text { - да } $$
$$ B(x) = (\Delta_1)^{-1} \cdot \delta(x) = (\alpha ^ {53})^{-1} \cdot 1 = \alpha ^ {10}; $$ 
$$ \delta(x)= \alpha ^ {53}x + 1 $$
$$ L = r - L = 1 - 0 = 1; $$
$$ r \neq 2t $$
$$ \text{Итерация } r = 2 $$
$$ \Delta_2 = \displaystyle\sum_{j = 0} ^ {L = 1} \delta_j S_{r - j} = \delta_0 S_{2 - 0} + \delta_1 S_{2 - 1} = 1 \cdot 1 + \alpha ^ {53} \cdot \alpha ^ {53} = 1 + \alpha ^ {43} = \alpha ^ {37}; $$
$$ \Delta_2 \neq 0 \implies T(x) = \delta(x) + \Delta_2 x B(x) = \alpha ^ {53} x + 1 + \alpha ^ {37} \cdot x \cdot \alpha ^ {10} = \alpha ^ {10}x + 1; $$
$$ 2 \cdot 1 \leqslant 2 - 1 $$
$$ 2 \leqslant 1 \text{ - нет} $$
$$ \delta(x)= \alpha ^ {10}x + 1 $$
$$ B(x) = x \cdot B(x) = \alpha ^ {10} x $$
$$ r \neq 2t $$
$$ \text{Итерация } r = 3 $$
$$ \Delta_3 = \displaystyle\sum_{j = 0} ^ {L = 1} \delta_j S_{r - j} = \delta_0 S_{3} + \delta_1 S_{2} = 1 \cdot \alpha ^ {10} + \alpha ^ {10} \cdot 1 = 0; $$
$$ \Delta_3 = 0 $$
$$ B(x) = x \cdot B(x) = \alpha ^ {10} x ^ 2 $$
$$ \text{Итерация } r = 4 $$
$$ \Delta_4 = \displaystyle\sum_{j = 0} ^ {L = 1} \delta_j S_{r - j} = \delta_0 S_{4} + \delta_1 S_{3} = 1 \cdot \alpha ^ {20} + \alpha ^ {10} \cdot \alpha ^ {10} = 0; $$
$$ \Delta_4 = 0 $$
$$ B(x) = x \cdot B(x) = \alpha ^ {10} x ^ 3 $$
$$ r = 2t \implies \text{Многочлен локаторов ошибок равен } \delta(x) = \alpha ^ {10}x + 1 \text{ , }L = 1;  $$
$$ \Omega(x)=(S(x)\delta(x))\mod(x ^ {2t}); $$
$$ \text{Вектор ошибки находится аналогично выше} $$
$$ \large{\text{Допустим две ошибки: }} $$
$$ U(x) = \alpha ^ {18} x ^ {12} + \alpha ^ {41} x ^ {9} + \underline{\alpha ^ {12} x ^ {8}} + \alpha ^ {54} x ^ {7} + \alpha ^ {6} x ^ {6} + \underline{\alpha ^ {38} x ^ {5}} + \alpha ^ {24} x ^ {4} + \alpha ^ {47} x ^ {3} + \alpha ^ {56} x ^ {2} + \alpha ^ {40} x + \alpha ^ {17}; $$
$$ S_1 = \alpha ^ {12} \cdot \alpha ^ {8} + \alpha ^ {38} \cdot \alpha ^ {5} = \alpha ^ {31}; $$
$$ S_2 = \alpha ^ {12} \cdot (\alpha ^ {2}) ^ {8} + \alpha ^ {38} \cdot (\alpha ^ {2}) ^ {5} = \alpha ^ {22} $$
$$ S_3 = \alpha ^ {12} \cdot (\alpha ^ {3}) ^ {8} + \alpha ^ {38} \cdot (\alpha ^ {3}) ^ {5} = \alpha ^ {12}; $$
$$ S_4 = \alpha ^ {12} \cdot (\alpha ^ {4}) ^ {8} + \alpha ^ {38} \cdot (\alpha ^ {4}) ^ {5} = \alpha ^ {60}; $$
$$ \underline{\text{Алгоритм Евклида: }} $$
$$ \text{Начальные значения: }a_0 = x ^ {2t} = x ^ 4, a_1 = b = S(x) = \displaystyle\sum_{j = 0} ^ {2t} S _ j x ^ {j - 1} = \alpha ^ {31} + \alpha ^ {22} x + \alpha ^ {12} x ^ 2 + \alpha ^ {60} x ^ 3, y_0 = 0, y_1 = 1 $$
$$ \text{При }i = 0\text{ : } $$
$$ a_0 = a_1 \cdot q_1 + a_2 \implies q_1 = \alpha ^ {3}x + (\alpha ^ {4} + \alpha ^ {2} + \alpha + 1) = \alpha ^ {3}x + \alpha ^ {18}; a_2 = \alpha ^ {55} x ^ 2 + \alpha ^ {60} x + \alpha ^ {49} $$
$$ \deg ( a_2 ) \leqslant t - 1 $$
$$ 2 \leqslant 1 \text{ - нет } $$
$$ y_2 = y_0 - y_1 \cdot q_1 = \alpha ^ {3} x + \alpha ^ {18};$$
$$ \text{При }i = 1\text{ : } $$
$$ a_1 = a_2 \cdot q_2 + a_3 \implies q_2 = \alpha ^ {5}x + (\alpha ^ {5} + \alpha ^ {4} + \alpha ^ {2} + \alpha) = \alpha ^ {5}x + \alpha ^ {7}; a_3 = (\alpha ^ {4} + \alpha ^ {3} + \alpha)x + (\alpha ^ {4} + \alpha ^ {3} + \alpha) = \alpha ^ {26}x + \alpha ^ {26} $$
$$ \deg ( a_3 ) \leqslant t - 1 $$
$$ 1 \leqslant 1 \text{ - да } \implies i = 3 $$
$$ y_3 = y_1 - y_2 \cdot q_2 = 1 + (\alpha ^ {3}x + \alpha ^ {18})(\alpha ^ {5}x + \alpha ^ {7}) = \alpha ^ {8} x ^ {2} + \alpha ^ {13} x + \alpha ^ {58}; $$
$$ \text{Найдем многочлен локаторов ошибок: } \delta(x) = \chi y_{i} = \chi y_{3} \text{ при условии, что } \delta(0) = 1\text { т.е подберем } \chi $$
$$ \chi = \alpha ^ {5} \text{ т.к. } \alpha ^ {5} (\alpha ^ {8} \cdot 0 ^ {2} + \alpha ^ {13} \cdot 0 + \alpha ^ {58})  = 1  $$
$$ \implies \delta(x) = \alpha ^ {5} (\alpha ^ {8} x ^ {2} + \alpha ^ {13} x + \alpha ^ {58}) =  $$
$$ = \alpha ^ {13} x ^ {2} + \alpha ^ {18} x + 1; $$
$$ \text{Корни многочлена локаторов ошибок равны обратным величинам локаторов ошибок} $$
$$ X_i = (\alpha ^ {j_1}) ^ {-1} \text{ - корень многочлена локаторов ошибок, а } j_i \text{ - место ошибки}$$
$$ \text{Корень } X_1 = \alpha ^ {55} \implies \text{ локатор ошибок равен } (\alpha ^ {55}) ^ {-1} = \alpha ^ {8} \implies j_1 = 8 \implies \text{ошибка в 8м месте} $$
$$ \text{Корень } X_1 = \alpha ^ {58} \implies \text{ локатор ошибок равен } (\alpha ^ {58}) ^ {-1} = \alpha ^ {5} \implies j_1 = 5 \implies \text{ошибка в 5м месте} $$
$$ \text{Многочлен значений значений ошибок имеет вид:} $$
$$ \Omega(x) = (-1) ^ {i} \chi a_i = (-1) ^ {i} \chi a_3 = $$
$$ = \alpha ^ {5} \cdot (\alpha ^ {26}x + \alpha ^ {26}) = \alpha ^ {31}x + \alpha ^ {31}; $$
$$ \text{Производная многочлена локаторов ошибок } \delta'(x) = \alpha ^ {18} $$
$$ \text{Теперь по алгоритму Форни получаем значения для вектора ошибок:} $$
$$ Y_i = - \frac{\Omega(X_i)}{\delta'(X_i)} $$
$$ Y_1 = - \frac{\Omega(X_1)}{\delta'(X_1)} = \frac{ \alpha ^ {31} \cdot \alpha ^ {55} + \alpha ^ {31}}{\alpha ^ {18}} =  $$
$$ = (\alpha ^ {18}) ^ {-1} \cdot (\alpha ^ {31 + 55} + \alpha ^ {31}) = \alpha ^ {12};  $$
$$ Y_2 = - \frac{\Omega(X_2)}{\delta'(X_2)} = \alpha ^ {45} \cdot (\alpha ^ {31} \cdot \alpha ^ {58} + \alpha ^ {31}) = \alpha ^ {38}  $$
$$ \implies \text{ вектор ошибок имеет вид } e(x) = \alpha ^ {12} x ^ {8} + \alpha ^ {38} x ^ {5}; $$
$$ \implies \text{ правильная комбинация определяется как } V(x) = U(x) + e(x); $$
$$ \underline{\text{Алгоритм Берлекэмпа-Месси: }} $$
$$ \text{Начальные условия: } r = 0, L = 0, \delta(x) = 1, B(x) = 1; $$
$$ \text{Итерация } r = r + 1 = 0 + 1 = 1 $$
$$ \Delta_1 = \displaystyle\sum_{j = 0} ^ {L = 0} \delta_j S_{r - j} = \alpha ^ {31}; $$
$$ \Delta_1 = \alpha ^ {31} \neq 0 \implies T(x) = \delta(x) - \Delta_r x B(x) = 1 + \alpha ^ {31} \cdot x \cdot 1 = \alpha ^ {31} x + 1; $$
$$ 2 \cdot 0 \leqslant 1 - 1 \text { - да } $$
$$ B(x) = (\Delta_r)^{-1} \cdot \delta(x) = \alpha ^ {32}; $$ 
$$ \delta(x)= \alpha ^ {31}x + 1 $$
$$ L = r - L = 1 - 0 = 1; $$
$$ \text{Итерация } r = 2 $$
$$ \Delta_2 = \displaystyle\sum_{j = 0} ^ {L = 1} \delta_0 S_{r - j} = \delta_0 S_2 + \delta_1 S_1 = 1 \cdot S_2 + \alpha ^ {31} \cdot S_1 = \alpha ^ {22} + \alpha ^ {31} \cdot \alpha ^ {31} = \alpha ^ {10} \neq 0 ; $$
$$ T(x) = \delta(x) + \Delta_2 \cdot x \cdot B(x) = (\alpha ^ {31} x + 1) + \alpha ^ {10} \cdot x \cdot \alpha ^ {32} = \alpha ^ {54}x + 1; $$
$$ 2 \cdot 1 \leqslant 2 - 1 \text { - нет } $$
$$ \delta(x)= \alpha ^ {54}x + 1; $$
$$ B(x) = x \cdot B(x) = \alpha ^ {32}x; $$ 
$$ \text{Итерация } r = 3 $$
$$ \Delta_3 = \displaystyle\sum_{j = 0} ^ {L = 1} \delta_j S_{r - j} = \delta_0 S_3 + \delta_1 S_2 = 1 \cdot \alpha ^ {12} + \alpha ^ {54} \cdot \alpha ^ {22} = \alpha ^ {5} \neq 0; $$
$$ T(x) = \delta(x) + \Delta_3 \cdot x \cdot B(x) = (\alpha ^ {54} x + 1) + \alpha ^ {5} \cdot x \cdot \alpha ^ {32} \cdot x = \alpha ^ {37}x ^ {2} + \alpha ^ {54}x + 1; $$
$$ 2 \cdot 1 \leqslant 3 - 1 \text { - да } $$
$$ B(x) = (\alpha ^ {5}) ^ {-1} \cdot (\alpha ^ {54}x + 1) = \alpha ^ {58}(\alpha ^ {54}x + 1) = \alpha ^ {49}x + \alpha ^ {58}; $$ 
$$ \delta(x) = \alpha ^ {37} x ^ {2} + \alpha ^ {54} x + 1; $$
$$ L = r - L = 3 - 1 = 2; $$
$$ \text{Итерация } r = 4 $$
$$ \Delta_4 = \displaystyle\sum_{j = 0} ^ {L = 2} \delta_j S_{r - j} = \delta_0 S_4 + \delta_1 S_3 + \delta_2 S_2 = S_4 + \alpha ^ {54} S_3 + \alpha ^ {37} S_2 = \alpha ^ {60} + \alpha ^ {54} \cdot \alpha ^ {12} \neq \alpha ^ {37} \cdot \alpha ^ {22} = \alpha ^ {5}; $$
$$ T(x) = \delta(x) + \Delta_5 \cdot x \cdot B(x) = (\alpha ^ {37} x ^ 2 + \alpha ^ {54} x + 1) + \alpha ^ {5} x \cdot (\alpha ^ {49} x + \alpha ^ {58}) = \alpha ^ {13} x ^ {2} + \alpha ^ {9} x + 1; $$
$$ 2 \cdot 2 \leqslant 4 - 1 \text { - нет } $$
$$ \implies \text{Многочлен локаторов ошибок равен } \delta(x) = \alpha ^ {13} x ^ 2 + \alpha ^ {18} x + 1; L = 2; $$
$$ \Omega(x)=(S(x)\delta(x))\mod(x ^ {2t}) = (\alpha ^ {60} x ^ {3} + \alpha ^ {12} x ^ {2} + \alpha ^ {22} x + \alpha ^ {31}) \cdot $$
$$ \cdot (\alpha ^ {13} x ^ {2} + \alpha ^ {18} x + 1) \mod (x ^ {4}) = (\alpha ^ {60 + 13} x ^ {5} + \alpha ^ {60 + 18} x ^ {4} +  $$
$$ + \alpha ^ {60} x ^ {3} + \alpha ^ {12 + 13} x ^ {4} + \alpha ^ {12 + 18} x ^ {3} + \alpha ^ {12} x ^ {2} + \alpha ^ {22 + 13} x ^ {3} + $$
$$ + \alpha ^ {22 + 18} x ^ {2} + \alpha ^ {22} x + \alpha ^ {31 + 13} x ^ {2} + \alpha ^ {31 + 18} x + \alpha ^ {31}) \mod (x ^ {4}) = $$
$$ = (\alpha ^ {10} x ^ {5} + \alpha ^ {15} x ^ {4} + \alpha ^ {60} x ^ {3} + \alpha ^ {25} x ^ {4} + \alpha ^ {30} x ^ {3} + $$
$$ + \alpha ^ {12} x ^ {2} + \alpha ^ {35} x ^ {3} + \alpha ^ {40} x ^ {2} + \alpha ^ {22} x + \alpha ^ {44} x ^ {2} + \alpha ^ {49} x + \alpha ^ {31}) \mod (x ^ {4}) $$
$$ = (\alpha ^ {10} x ^ {5} + (\alpha ^ {15} + \alpha ^ {25}) x ^ {4} + (\alpha ^ {60} + \alpha ^ {30} + \alpha ^ {35}) x ^ {3} + $$
$$ + (\alpha ^ {12} + \alpha ^ {40} + \alpha ^ {44}) x ^ {2} + (\alpha ^ {22} + \alpha ^ {49}) x + \alpha ^ {31})\mod (x ^ {4}) = $$
$$ = (\alpha ^ {10} x ^ 5 + \alpha ^ {12} x ^ 4 + \alpha ^ {31} x + \alpha ^ {31})\mod (x ^ {4}) = \alpha ^ {31} x + \alpha ^ {31} $$
$$ \text{Вектор ошибки находится аналогично выше} $$
